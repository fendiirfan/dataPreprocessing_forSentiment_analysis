```
import string

def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
#start process_tweet
# def repair_repeat_alphabet(kalimat):
#   kalimat = kalimat.split()
#   for index2 in range(len(kalimat)):
#     # print(kalimat[index2])
#     for index3 in range(len(kalimat[index2])):
#       # print('index2')
#       # print(index2)
#       # print('index3')
#       # print(index3)
#       if index3 <= len(kalimat[index2])-3:
#         # print(kalimat[index2])
#         if kalimat[index2][index3] == kalimat[index2][index3+2]:
#           # print(kalimat[index2])
#           # print(kalimat[index2][index3])
#           huruf = kalimat[index2][index3]
#           temp = huruf
#           i = index3
#           print(kalimat[index2])
#           print(len(kalimat[index2]))
#           while huruf == kalimat[index2][i] and i <= len(kalimat[index2])-2:
#             # print(kalimat[index2][i])
#             temp += kalimat[index2][index3]
#             # print(temp)
#             i += 1
#           if huruf != kalimat[index2][len(kalimat[index2])-1]:
#             temp = temp[:len(temp)-1]
#           kalimat = ' '.join(kalimat)
#           # print(temp)
#           # print(huruf)
#           # kamus['']
#           kalimat = kalimat.replace(temp,huruf)
#           kalimat = kalimat.split()
#   kalimat = ' '.join(kalimat)
#   return kalimat

def processTweet(tweet):
    # process the tweets

    #Convert to lower case
    tweet = tweet.lower()
    #Convert www.* or https?://* to URL
    tweet = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','URL',tweet)
    #Convert @username to AT_USER
    tweet = re.sub('@[^\s]+','AT_USER',tweet)
    #Remove additional white spaces
    tweet = re.sub('[\s]+', ' ', tweet)
    #Replace #word with word
    tweet = re.sub(r'#([^\s]+)', r'\1', tweet)
    #trim
    tweet = tweet.strip('\'"')
    # deleted punctuation 
    tweet = tweet.translate(str.maketrans('','',string.punctuation)).lower()
    # deleted emojis
    tweet = remove_emoji(tweet)
    # deleted pattern pictwitter
    tweet = re.sub('pictwitter.+',"",tweet)
    # deleted pattern url
    tweet = re.sub('url',"",tweet)
    # repeat aphabet
    # tweet = repair_repeat_alphabet(tweet)
    
    return tweet


PreprocessedTextDf["text"] = PreprocessedTextDf["text"].apply(processTweet) # menjalankan fungsi processTweet untuk melakukan bebrapa tugas pembersihan data


```

```
def perbaikan_huruf_berulang(kalimat):
  kalimat = kalimat.split()
  for index2 in range(len(kalimat)):
    # print(kalimat[index2])
    for index3 in range(len(kalimat[index2])):
      # print('index2')
      # print(index2)
      # print('index3')
      # print(index3)
      if index3 <= len(kalimat[index2])-3:
        # print(kalimat[index2])
        if kalimat[index2][index3] == kalimat[index2][index3+2]:
          # print(kalimat[index2])
          # print(kalimat[index2][index3])
          huruf = kalimat[index2][index3]
          temp = huruf
          i = index3
          while huruf == kalimat[index2][i] and i <= len(kalimat[index2])-2:
            # print(kalimat[index2][i])
            temp += kalimat[index2][index3]
            # print(temp)
            i += 1
          if huruf != kalimat[index2][len(kalimat[index2])-1]:
            temp = temp[:len(temp)-1]
          kalimat = ' '.join(kalimat)
          # print(temp)
          # print(huruf)
          # kamus['']
          kalimat = kalimat.replace(temp,huruf)
          kalimat = kalimat.split()
  kalimat = ' '.join(kalimat)
  return kalimat

for i in data_used:
  PreprocessedTextDf['text'][i] = perbaikan_huruf_berulang(PreprocessedTextDf['text'][i])
  PreprocessedTextDf['text'][i] = PreprocessedTextDf['text'][i].replace('²','')
  PreprocessedTextDf['text'][i] = PreprocessedTextDf['text'][i].replace('2','')
```
